```python
citibike_df.drop('birth year', axis=1, inplace=True)
citibike_df.head()
```

```python
citibike_df['starttime'] = pd.to_datetime(citibike_df["starttime"])
citibike_df['stoptime'] = pd.to_datetime(citibike_df['stoptime'])
```

```python
citibike_df.head()
```

```python
citibike_df['trip duration'] = citibike_df['stoptime'] - citibike_df['starttime']
citibike_df['trip duration'].head()
```

```python
def get_weekend(weekday):
    return int(weekday >= 5)
```

```python
citibike_df['weekend'] = citibike_df['starttime'].dt.dayofweek.apply(get_weekend)
```

```python
citibike_df['weekend'].value_counts()
```

```python
def get_timeofday(time):
    if 0<=time<=6:
        return "night"
    elif 6<time<=12:
        return 'morning'
    elif 12<time<=18:
        return "day"
    elif 18<time<=23:
        return "evening"
    
```

```python
citibike_df['time_of_day'] = citibike_df['starttime'].dt.hour.apply(get_timeofday)
citibike_df.head()
```

```python
display(citibike_df.loc[:, ['starttime', 'time_of_day']])
```

```python
day_time = citibike_df[citibike_df['time_of_day'] == 'day'].shape[0]
night_time = citibike_df[citibike_df['time_of_day'] == 'night'].shape[0]
print(day_time, night_time, day_time / night_time)
```
